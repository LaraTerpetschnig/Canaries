# Cleaning map the meal gap dataset

```python
drop_columns = ['FIPS', 'Year', 'Rural-Urban Continuum Code (2023)', 'Census Region', 'Census Division']
meal_gap_2019 = meal_gap_2019.drop(drop_columns, axis=1)
```

```python
meal_gap_2019.shape
```

```python
meal_gap_2019_columns = ['State', 
                         'County, State',
                         'pctFoodInsecure',
                         'numFoodInsecure',
                         'pctBlackFoodInsecure',
                         'pctLatineFoodInsecure',
                         'pctWhiteFoodInsecure',
                         'SNAPthreshold',
                         'pctFIbelowSNAP',
                         'pctFIaboveSNAP',
                         'pctKidsFoodInsecure',
                         'numKidsFoodInsecure',
                         'pctKidsIncomeBelow185',
                         'pctKidsIncomeAbove185',
                         'costPerMeal',
                         'weeklyFoodBudget',
                         'weeklyBudgetFoodInsecureByPop',
                         'rangeRuralUrban',
                         'USstateRegion']
```

```python
len(meal_gap_2019_columns)
```

```python
meal_gap_2019 = meal_gap_2019.set_axis(meal_gap_2019_columns, axis=1)
meal_gap_2019.head()
```

```python
meal_gap_2019['pctFoodInsecure'] = meal_gap_2019['pctFoodInsecure'].str[:-1].astype('float')
meal_gap_2019['pctBlackFoodInsecure'] = meal_gap_2019['pctBlackFoodInsecure'].str[:-1].astype('float')
meal_gap_2019['pctLatineFoodInsecure'] = meal_gap_2019['pctLatineFoodInsecure'].str[:-1].astype('float')
meal_gap_2019['pctWhiteFoodInsecure'] = meal_gap_2019['pctWhiteFoodInsecure'].str[:-1].astype('float')
meal_gap_2019['pctKidsFoodInsecure'] = meal_gap_2019['pctKidsFoodInsecure'].str[:-1].astype('float')


meal_gap_2019['SNAPthreshold'] = meal_gap_2019['SNAPthreshold'].str[:-1].astype('int')
meal_gap_2019['pctFIbelowSNAP'] = meal_gap_2019['pctFIbelowSNAP'].str[:-1].astype('int')
meal_gap_2019['pctFIaboveSNAP'] = meal_gap_2019['pctFIaboveSNAP'].str[:-1].astype('int')
meal_gap_2019['pctKidsIncomeBelow185'] = meal_gap_2019['pctKidsIncomeBelow185'].str[:-1].astype('int')
meal_gap_2019['pctKidsIncomeAbove185'] = meal_gap_2019['pctKidsIncomeAbove185'].str[:-1].astype('int')

meal_gap_2019['rangeRuralUrban'] = meal_gap_2019['rangeRuralUrban'].astype('int')
```

```python
meal_gap_2019['numFoodInsecure'] = meal_gap_2019['numFoodInsecure'].apply(lambda x : int(x.replace(",", "")) if pd.notna(x) else x)
meal_gap_2019['numKidsFoodInsecure'] = meal_gap_2019['numKidsFoodInsecure'].apply(lambda x : int(x.replace(",", "")) if pd.notna(x) else x)
```

```python
meal_gap_2019['costPerMeal'] = meal_gap_2019['costPerMeal'].str[1:].astype('float')
meal_gap_2019['weeklyFoodBudget'] = meal_gap_2019['weeklyFoodBudget'].str[1:].astype('float')

meal_gap_2019['weeklyBudgetFoodInsecureByPop'] = meal_gap_2019['weeklyBudgetFoodInsecureByPop'].str[1:].apply(lambda x : int(x.replace(",", "")) if pd.notna(x) else x)
```
