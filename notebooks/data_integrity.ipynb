```python
import hashlib
```

```python
with open('../data/MMG2025_2019-2023_Data_To_Share.xlsx', "rb") as f:
    file = f.read()

hex_digest = hashlib.sha256(file).hexdigest()

with open('../hash digest/mmg.sha', "w") as f:
    f.write(hex_digest)
```
