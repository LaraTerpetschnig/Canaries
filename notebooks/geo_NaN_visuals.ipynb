```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
```

## Load data

```python
df = pd.read_csv("../data/MMG_FARA_2019_county.csv", dtype={'FIPS': str})
df['FIPS'] = df['FIPS'].str.zfill(5)

counties = gpd.read_file('https://www2.census.gov/geo/tiger/GENZ2019/shp/cb_2019_us_county_20m.zip')
counties['FIPS'] = counties['STATEFP'] + counties['COUNTYFP']

contig = counties[~counties['STATEFP'].isin(['02', '15', '60', '66', '69', '72', '78'])]

gdf = contig.merge(df, on='FIPS', how='left').to_crs('ESRI:102003')

if 'LILATracts_1And10' in gdf.columns and 'n_tracts' in gdf.columns:
    gdf['lila_share'] = gdf['LILATracts_1And10'] / gdf['n_tracts']
```

# NaN Value Visualizations

## laseniors20

```python
key_vars = [
    'laseniors20'
]

gdf['nan_count'] = gdf[key_vars].isna().sum(axis=1)
gdf['any_missing'] = gdf['nan_count'] > 0

fig, ax = plt.subplots(figsize=(12, 8))

gdf.plot(
    column='nan_count',
    ax=ax,
    cmap='Reds',
    legend=True,
    legend_kwds={'shrink': 0.6, 'label': 'Number of missing variables'},
    edgecolor='white',
    linewidth=0.1,
)

ax.set_title('Missing laseniors20 Data (2019)', fontsize=14)
ax.axis('off')
plt.tight_layout()
plt.savefig('../visualizations/missing_laseniors20_data_map.png', dpi=150, bbox_inches='tight')
plt.show()
```
