```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("food_insecurity_merged.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Urban', color='burlywood')
plt.title('Counties by Urban/Rural Status of their Census Tracts')
plt.xlabel('Urban Flag Average (0=Rural, 1=Urban)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df_state_insecurity = df[['State', 'pctFoodInsecure']].groupby('State').agg('mean').reset_index().sort_values(by='pctFoodInsecure', ascending=False).iloc[:20]
df_state_insecurity.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_state_insecurity, x='State', y='pctFoodInsecure', color='burlywood')
plt.title('Average % of Food Insecure Individuals per State')
plt.xlabel('State')
plt.ylabel('% Food Insecure')
plt.tight_layout()
plt.show()
```
